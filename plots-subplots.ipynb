```python
from aim2dat.plots import SimplePlot

splot = SimplePlot()
splot.import_scatter_data_set(
    "test data set", [0.0, 1.0, 2.0], [0.0, 2.0, 5.0], color="C0"
)
splot.import_bar_data_set(
    "test data set 2", 
    [0.0, 1.0, 2.0], 
    [0.0, 2.0, 5.0], 
    [2.0, 4.0, 2.0], 
    color="C1"
)
splot.import_scatter_data_set(
    "test data set 3", 
    [-1.0, 0.0, 2.0, 3.0], 
    [-5.0, 2.0, 3.0, -1.0], 
    color="C2"
)
```

```python
splot.subplot_nrows = 2
splot.subplot_ncols = 2
splot.plot(
    ["test data set", "test data set 2", "test data set 3"], 
    subplot_assignment=[1, 2, 0]
)
```

```python
splot.create_default_gridspec(
    nplots=3, nrows=2, ncols=2, center_last_row=True
)
splot.subplot_wspace = 0.5
splot.plot(
    ["test data set", "test data set 2", "test data set 3"], 
    subplot_assignment=[1, 2, 0]
)
```

```python
splot.reset_gridspec()
splot.plot(
    ["test data set", "test data set 2", "test data set 3"], 
    subplot_assignment=[1, 2, 0]
)
```
